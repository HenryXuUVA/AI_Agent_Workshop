```python
%%writefile llama_eval.py
import argparse
import torch
import time
import json
import os
import warnings
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from datasets import load_dataset
from tqdm import tqdm

# Filter warnings
warnings.filterwarnings("ignore")

MODELS = [
    'meta-llama/Llama-3.2-3B-Instruct',
    'Qwen/Qwen2.5-1.5B-Instruct'
]

SUBJECTS = [
    'astronomy',
    'business_ethics',
    'elementary_mathematics',
    'marketing',
    'global_facts',
    'machine_learning',
    'philosophy',
    'professional_law',
    'world_religions',
    'virology'
]

def format_subject(subject):
    l = subject.split("_")
    s = ""
    for entry in l:
        s += " " + entry
    return s

def format_example(df, idx, include_answer=True):
    prompt = df['question'][idx]
    k = df.shape[1] - 2
    for j in range(k):
        prompt += "\n{}. {}".format(chr(65 + j), df['choices'][idx][j])
    prompt += "\nAnswer:"
    if include_answer:
        prompt += " {}\n\n".format(chr(65 + df['answer'][idx]))
    return prompt

def gen_prompt(train_df, subject, k=-1):
    prompt = "The following are multiple choice questions (with answers) about {}.\n\n".format(
        format_subject(subject)
    )
    if k == -1:
        k = train_df.shape[0]
    for i in range(k):
        prompt += format_example(train_df, i)
    return prompt

def evaluate_subject(model, tokenizer, subject_name, device, verbose=False):
    try:
        # Load dataset
        dataset = load_dataset("cais/mmlu", subject_name, split="test")

        # Simple 5-shot prompt construction (using dev set if available, else zero-shot or just instructions)
        # For simplicity and speed in this benchmark, we'll use 0-shot or very basic instruction
        # typically MMLU uses 5-shot. Let's try to load dev for shots.
        try:
            dev_df = load_dataset("cais/mmlu", subject_name, split="dev")
            base_prompt = gen_prompt(dev_df, subject_name, k=5)
        except:
            base_prompt = "The following are multiple choice questions (with answers) about {}.\n\n".format(format_subject(subject_name))

        correct = 0
        total = 0

        # Synchronize GPU before starting timer if applicable
        if device == "cuda":
            torch.cuda.synchronize()
            start_event = torch.cuda.Event(enable_timing=True)
            end_event = torch.cuda.Event(enable_timing=True)
            start_event.record()

        start_time_real = time.time()
        start_time_cpu = time.process_time()

        for i, example in enumerate(tqdm(dataset, desc=f"Evaluating {subject_name}", leave=False)):
            # Prepare prompt
            prompt_end = format_example(dataset, i, include_answer=False)
            full_prompt = base_prompt + prompt_end

            inputs = tokenizer(full_prompt, return_tensors="pt").to(device)

            # Generate
            with torch.no_grad():
                outputs = model.generate(
                    **inputs,
                    max_new_tokens=1,
                    pad_token_id=tokenizer.eos_token_id
                )

            # Decode
            output_token = tokenizer.decode(outputs[0, -1]).strip()

            # Check answer
            label = chr(65 + example['answer'])
            is_correct = output_token.upper() == label
            if is_correct:
                correct += 1
            total += 1

            if verbose:
                print(f"\nQ: {example['question'][:100]}...")
                print(f"Pred: {output_token}, Label: {label}, Correct: {is_correct}")

        # End timing
        end_time_real = time.time()
        end_time_cpu = time.process_time()

        gpu_time = 0.0
        if device == "cuda":
            end_event.record()
            torch.cuda.synchronize()
            gpu_time = start_event.elapsed_time(end_event) / 1000.0 # Convert ms to s

        real_time = end_time_real - start_time_real
        cpu_time = end_time_cpu - start_time_cpu

        return {
            "accuracy": correct / total if total > 0 else 0,
            "correct": correct,
            "total": total,
            "time_real": real_time,
            "time_cpu": cpu_time,
            "time_gpu": gpu_time
        }

    except Exception as e:
        print(f"Error evaluating {subject_name}: {e}")
        return None

def main():
    parser = argparse.ArgumentParser()
    parser.add_argument("--verbose", action="store_true", help="Enable verbose logging")
    parser.add_argument("--device", type=str, default="cuda" if torch.cuda.is_available() else "cpu")
    parser.add_argument("--quantization", type=str, choices=["none", "4bit", "8bit"], default="none")
    args = parser.parse_args()

    results = {}

    for model_name in MODELS:
        print(f"\nLoading model: {model_name}...")

        # Config for quantization
        bnb_config = None
        if args.quantization == "4bit":
            bnb_config = BitsAndBytesConfig(
                load_in_4bit=True,
                bnb_4bit_compute_dtype=torch.float16,
                bnb_4bit_quant_type="nf4"
            )
        elif args.quantization == "8bit":
            bnb_config = BitsAndBytesConfig(load_in_8bit=True)

        try:
            tokenizer = AutoTokenizer.from_pretrained(model_name)
            # Set pad token if missing
            if tokenizer.pad_token is None:
                tokenizer.pad_token = tokenizer.eos_token

            if args.device == "cuda" and args.quantization != "none":
                model = AutoModelForCausalLM.from_pretrained(
                    model_name,
                    quantization_config=bnb_config,
                    device_map="auto",
                    trust_remote_code=True
                )
            else:
                model = AutoModelForCausalLM.from_pretrained(
                    model_name,
                    device_map=args.device if args.device == "cuda" else None,
                    trust_remote_code=True
                )
                if args.device == "cpu":
                    model = model.to("cpu")

            model.eval()

            model_results = {}
            for subject in SUBJECTS:
                print(f"Evaluating {subject} on {model_name}...")
                res = evaluate_subject(model, tokenizer, subject, args.device, args.verbose)
                if res:
                    model_results[subject] = res
                    print(f"  Accuracy: {res['accuracy']:.2%} | Time (Real): {res['time_real']:.2f}s")

            results[model_name] = model_results

            # Clear memory
            del model
            del tokenizer
            torch.cuda.empty_cache()

        except Exception as e:
            print(f"Failed to load or evaluate {model_name}: {e}")

    # Save results
    with open("multi_model_benchmark_results.json", "w") as f:
        json.dump(results, f, indent=2)

    print("\nBenchmark complete. Results saved to multi_model_benchmark_results.json")

if __name__ == "__main__":
    main()
```
